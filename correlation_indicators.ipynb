```python
raw_data=pd.read_csv("./"+"final Urban Development.csv")
gdp_raw_data=pd.read_csv("./GDP.csv",skiprows=3)
raw_data=raw_data.drop(['Unnamed: 0'], axis=1)
raw_data
raw_data.columns  = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2003', '2004','2005', '2006', '2007', '2008', '2009','2010', '2011', '2012', '2013', '2014','2015', '2016','2017',2018]
raw_data.head()
```

```python
def dataProcessing(df, year):
    dfAggeregateAll = pd.DataFrame(columns = ['Country Name','Year'])
    k = 1  
    df = df[['Country Name','Indicator Name', year]] 
    listIndicatorCode = list(set(list(np.unique(df["Indicator Name"])))) 
    for j in listIndicatorCode:
        df_sub = df.loc[df['Indicator Name'] == j]
        df_sub['Year'] = year
        df_sub[j] = df_sub[year]
        del df_sub[year]
        del df_sub['Indicator Name']
        dfAggeregateAll = dfAggeregateAll.drop_duplicates() 
        dfAggeregateAll = pd.merge(dfAggeregateAll,df_sub, how = "outer", on = ['Country Name', 'Year'])
    dfAggeregateAll = dfAggeregateAll.drop_duplicates()       
    return dfAggeregateAll 
```

```python
dfAllYear=pd.DataFrame()
for year in range(2003,2017):
    dfAggeregateAll = dataProcessing(raw_data,str(year))
    dfAllYear = dfAllYear.append(dfAggeregateAll)
    print(year)
```
